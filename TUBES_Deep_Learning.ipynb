```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping

# Set random seeds untuk reproducibility
np.random.seed(42)
import tensorflow as tf
tf.random.set_seed(42)

# Membaca dataset
print("üìä Memuat dataset...")
df = pd.read_csv('Dataset1.csv')

# Mengecek informasi dataset
print(f"\nüìã Informasi dataset: {df.shape[0]} baris √ó {df.shape[1]} kolom")
print("\nüîç Sampel data:")
print(df.head(5))

# Mengecek jumlah nilai NaN per kolom
print("\nüîé Jumlah NaN sebelum pengisian:")
print(df.isna().sum())

# Mengisi nilai NaN dengan rata-rata untuk kolom numerik
df.fillna(df.mean(numeric_only=True), inplace=True)

# Mengecek kembali jumlah NaN setelah pengisian
print("\n‚úÖ Jumlah NaN setelah pengisian:")
print(df.isna().sum().sum())

# Visualisasi distribusi target
plt.figure(figsize=(10, 6))
sns.countplot(x=df['churn'])
plt.title('Distribusi Churn', fontsize=16)
plt.xlabel('Churn Status', fontsize=14)
plt.ylabel('Jumlah Pelanggan', fontsize=14)
plt.tight_layout()
plt.show()

# Encode kolom target "churn" (Yes/No ‚Üí 1/0)
print("\nüîÑ Encoding target variable...")
le = LabelEncoder()
df['churn'] = le.fit_transform(df['churn'])
print(f"Label encoding: {le.classes_} ‚Üí {le.transform(le.classes_)}")

# Feature correlation analysis
print("\nüìà Analisis korelasi fitur dengan target...")
plt.figure(figsize=(12, 10))
correlation = df.corr()
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='coolwarm',
            linewidths=0.5, vmin=-1, vmax=1)
plt.title('Korelasi Antar Fitur', fontsize=16)
plt.tight_layout()
plt.show()

# Top features correlation with churn
churn_corr = correlation['churn'].drop('churn').sort_values(ascending=False)
print("\nTop korelasi dengan churn:")
print(churn_corr.head(5))

# Pisahkan fitur dan target
print("\nüîÑ Mempersiapkan data untuk pemodelan...")
X = df.drop("churn", axis=1)
y = df["churn"]

# One-hot encode kategori
categorical_cols = X.select_dtypes(include=['object']).columns
if len(categorical_cols) > 0:
    print(f"One-hot encoding kolom kategori: {list(categorical_cols)}")
    X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# Normalisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(f"Data dinormalisasi dengan StandardScaler. Shape: {X_scaled.shape}")

# Train-test split dengan mempertahankan index untuk tracking
X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
    X_scaled, y, df.index, test_size=0.3, random_state=42, stratify=y)

print(f"Train set: {X_train.shape[0]} sampel, Test set: {X_test.shape[0]} sampel")

# Reshape data untuk CNN
n_features = X_train.shape[1]
print(f"Jumlah fitur: {n_features}")

# Find best shape for reshaping (trying to make it square-like)
best_diff = float('inf')
for i in range(1, int(np.sqrt(n_features)) + 1):
    if n_features % i == 0:
        diff = abs(i - n_features // i)
        if diff < best_diff:
            best_diff = diff
            rows, cols = i, n_features // i

# If no exact factors found, pad the data
if best_diff == float('inf'):
    print("Tidak dapat menemukan faktor yang tepat. Menambahkan padding...")
    sqrt_features = int(np.ceil(np.sqrt(n_features)))
    rows = cols = sqrt_features
    padding = (sqrt_features * sqrt_features) - n_features

    # Add padding columns
    X_train = np.pad(X_train, ((0, 0), (0, padding)), 'constant')
    X_test = np.pad(X_test, ((0, 0), (0, padding)), 'constant')

    print(f"Menambahkan {padding} kolom padding. Shape baru: {X_train.shape}")

print(f"Reshape ke ({rows}, {cols}, 1)")
X_train_cnn = X_train.reshape(X_train.shape[0], rows, cols, 1)
X_test_cnn = X_test.reshape(X_test.shape[0], rows, cols, 1)
input_shape = (rows, cols, 1)

# Class balancing
print("\n‚öñÔ∏è Menghitung class weights untuk mengatasi imbalance...")
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
class_weights_dict = {0: class_weights[0], 1: class_weights[1]}
print(f"Class weights: {class_weights_dict}")

# CNN Model (Simplified)
print("\nüß† Membangun arsitektur CNN model...")
model = models.Sequential([
    layers.Conv2D(32, (2, 2), activation='relu', input_shape=input_shape),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(1, activation='sigmoid')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Model summary
model.summary()

# Setup early stopping
early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=10,
    restore_best_weights=True,
    verbose=1
)

# Training
print("\nüèãÔ∏è Melatih model CNN...")
history = model.fit(
    X_train_cnn, y_train,
    epochs=100,
    batch_size=64,
    validation_split=0.2,
    class_weight=class_weights_dict,
    callbacks=[early_stopping],
    verbose=1
)

# Visualisasi training history
plt.figure(figsize=(16, 6))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Akurasi Model')
plt.xlabel('Epoch')
plt.ylabel('Akurasi')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Loss Model')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Evaluation on test set
print("\nüìä Evaluasi model pada test set...")
test_loss, test_acc = model.evaluate(X_test_cnn, y_test, verbose=1)
print(f"Test loss: {test_loss:.4f}")
print(f"Test accuracy: {test_acc:.4f}")

# Prediksi probabilitas
y_pred_prob = model.predict(X_test_cnn)
# Konversi ke kelas 0/1
y_pred_class = (y_pred_prob > 0.5).astype(int)

# Metrics calculation
accuracy = accuracy_score(y_test, y_pred_class)
cm = confusion_matrix(y_test, y_pred_class)

# Print detailed metrics
print("\nüìà Metrik Performa Model:")
print(f"Accuracy: {accuracy:.4f}")
print("\nüìä Confusion Matrix:")
print(cm)

print("\nüìù Classification Report:")
print(classification_report(y_test, y_pred_class))

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)
print(f"AUC: {roc_auc:.4f}")

# Enhanced visualization - Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Tidak Churn", "Churn"],
            yticklabels=["Tidak Churn", "Churn"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()

# Plot ROC Curve
plt.figure(figsize=(10, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()

# Calculate optimal threshold
j_scores = tpr - fpr
optimal_idx = np.argmax(j_scores)
optimal_threshold = thresholds[optimal_idx]
print(f"\nüéØ Optimal threshold: {optimal_threshold:.4f}")

# Recalculate predictions with optimal threshold
y_pred_optimal = (y_pred_prob >= optimal_threshold).astype(int)
cm_optimal = confusion_matrix(y_test, y_pred_optimal)
accuracy_optimal = accuracy_score(y_test, y_pred_optimal)

print("\nüìä Metrik dengan threshold optimal:")
print(f"Accuracy: {accuracy_optimal:.4f}")
print(f"Confusion Matrix:\n{cm_optimal}")
print(classification_report(y_test, y_pred_optimal))

# Business metrics
total_customers = len(y_test)
actual_churn = np.sum(y_test)
predicted_churn = np.sum(y_pred_class)
predicted_churn_optimal = np.sum(y_pred_optimal)

print("\nüíº Metrik Bisnis:")
print(f"Total pelanggan dalam test set: {total_customers}")
print(f"Pelanggan aktual yang churn: {actual_churn} ({actual_churn/total_customers:.2%})")
print(f"Pelanggan diprediksi churn (threshold=0.5): {predicted_churn} ({predicted_churn/total_customers:.2%})")
print(f"Pelanggan diprediksi churn (threshold optimal={optimal_threshold:.4f}): {predicted_churn_optimal} ({predicted_churn_optimal/total_customers:.2%})")

# Identify high-risk customers
hasil_df = pd.DataFrame({
    'id': idx_test,
    'actual_churn': y_test,
    'predicted_churn': y_pred_optimal.flatten(),
    'churn_probability': y_pred_prob.flatten().round(4)
})

# Sort by churn probability (highest first)
high_risk_customers = hasil_df.sort_values('churn_probability', ascending=False).head(10)
print("\n‚ö†Ô∏è 10 Pelanggan dengan risiko churn tertinggi:")
print(high_risk_customers)

# Save prediction results to CSV
hasil_df.to_csv("hasil_prediksi_churn.csv", index=False)
print("\n‚úÖ Hasil prediksi telah disimpan ke 'hasil_prediksi_churn.csv'")

# Save the model
model.save('telecom_churn_cnn_model.h5')
print("‚úÖ Model telah disimpan ke 'telecom_churn_cnn_model.h5'")

# Create a brief summary report
print("\nüìã Ringkasan Model Prediksi Churn")
print("=" * 50)
print(f"Jumlah total pelanggan dalam analisis: {len(df)}")
print(f"Jumlah total pelanggan churn: {df['churn'].sum()} ({df['churn'].mean():.2%})")
print(f"Akurasi model: {accuracy:.2%}")
print(f"AUC score: {roc_auc:.4f}")
print("\n‚úÖ Analisis Selesai!")
```

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam

# Set random seeds untuk reproducibility
np.random.seed(42)
import tensorflow as tf
tf.random.set_seed(42)

# Membaca dataset
print("üìä Memuat dataset...")
df = pd.read_csv('Dataset1.csv')

# Mengecek informasi dataset
print(f"\nüìã Informasi dataset: {df.shape[0]} baris √ó {df.shape[1]} kolom")
print("\nüîç Sampel data:")
print(df.head(5))

# Mengecek jumlah nilai NaN per kolom
print("\nüîé Jumlah NaN sebelum pengisian:")
print(df.isna().sum())

# Mengisi nilai NaN dengan rata-rata untuk kolom numerik
df.fillna(df.mean(numeric_only=True), inplace=True)

# Mengecek kembali jumlah NaN setelah pengisian
print("\n‚úÖ Jumlah NaN setelah pengisian:")
print(df.isna().sum().sum())

# Visualisasi distribusi target
plt.figure(figsize=(10, 6))
sns.countplot(x=df['churn'])
plt.title('Distribusi Churn', fontsize=16)
plt.xlabel('Churn Status', fontsize=14)
plt.ylabel('Jumlah Pelanggan', fontsize=14)
plt.tight_layout()
plt.show()

# Encode kolom target "churn" (Yes/No ‚Üí 1/0) jika belum berupa angka
print("\nüîÑ Encoding target variable...")
if df['churn'].dtype == 'object':
    le = LabelEncoder()
    df['churn'] = le.fit_transform(df['churn'])
    print(f"Label encoding: {le.classes_} ‚Üí {le.transform(le.classes_)}")

# Feature correlation analysis
print("\nüìà Analisis korelasi fitur dengan target...")
plt.figure(figsize=(12, 10))
correlation = df.corr()
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='coolwarm',
            linewidths=0.5, vmin=-1, vmax=1)
plt.title('Korelasi Antar Fitur', fontsize=16)
plt.tight_layout()
plt.show()

# Top features correlation with churn
churn_corr = correlation['churn'].drop('churn').sort_values(ascending=False)
print("\nTop korelasi dengan churn:")
print(churn_corr.head(5))

# Pisahkan fitur dan target
print("\nüîÑ Mempersiapkan data untuk pemodelan...")
X = df.drop("churn", axis=1)  # Tidak menghapus 'id' seperti pada CNN untuk konsistensi
y = df["churn"]

# One-hot encode kategori
categorical_cols = X.select_dtypes(include=['object']).columns
if len(categorical_cols) > 0:
    print(f"One-hot encoding kolom kategori: {list(categorical_cols)}")
    X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# Normalisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(f"Data dinormalisasi dengan StandardScaler. Shape: {X_scaled.shape}")

# Train-test split dengan mempertahankan index untuk tracking
X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
    X_scaled, y, df.index, test_size=0.3, random_state=42, stratify=y)

print(f"Train set: {X_train.shape[0]} sampel, Test set: {X_test.shape[0]} sampel")

# Class balancing
print("\n‚öñÔ∏è Menghitung class weights untuk mengatasi imbalance...")
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
class_weights_dict = {0: class_weights[0], 1: class_weights[1]}
print(f"Class weights: {class_weights_dict}")

# Enhanced MLP Model
print("\nüß† Membangun arsitektur MLP model yang ditingkatkan...")
input_dim = X_train.shape[1]
model = models.Sequential([
    layers.Dense(256, activation='relu', input_dim=input_dim, kernel_regularizer=regularizers.l2(0.001)),
    layers.BatchNormalization(),
    layers.Dropout(0.3),

    layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    layers.BatchNormalization(),
    layers.Dropout(0.3),

    layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    layers.BatchNormalization(),
    layers.Dropout(0.2),

    layers.Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    layers.BatchNormalization(),
    layers.Dropout(0.1),

    layers.Dense(1, activation='sigmoid')
])

# Compile model dengan learning rate yang dikustomisasi
optimizer = Adam(learning_rate=0.001)
model.compile(
    optimizer=optimizer,
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Model summary
model.summary()

# Setup callbacks
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=5,
    min_lr=0.00001,
    verbose=1
)

# Training
print("\nüèãÔ∏è Melatih model MLP...")
history = model.fit(
    X_train, y_train,
    epochs=150,
    batch_size=32,
    validation_split=0.2,
    class_weight=class_weights_dict,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

# Visualisasi training history
plt.figure(figsize=(16, 6))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Akurasi Model')
plt.xlabel('Epoch')
plt.ylabel('Akurasi')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Loss Model')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Evaluation on test set
print("\nüìä Evaluasi model pada test set...")
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=1)
print(f"Test loss: {test_loss:.4f}")
print(f"Test accuracy: {test_acc:.4f}")

# Prediksi probabilitas
y_pred_prob = model.predict(X_test)
# Konversi ke kelas 0/1
y_pred_class = (y_pred_prob > 0.5).astype(int)

# Metrics calculation
accuracy = accuracy_score(y_test, y_pred_class)
cm = confusion_matrix(y_test, y_pred_class)

# Print detailed metrics
print("\nüìà Metrik Performa Model:")
print(f"Accuracy: {accuracy:.4f}")
print("\nüìä Confusion Matrix:")
print(cm)

print("\nüìù Classification Report:")
print(classification_report(y_test, y_pred_class))

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)
print(f"AUC: {roc_auc:.4f}")

# Enhanced visualization - Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Tidak Churn", "Churn"],
            yticklabels=["Tidak Churn", "Churn"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()

# Plot ROC Curve
plt.figure(figsize=(10, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()

# Calculate optimal threshold
j_scores = tpr - fpr
optimal_idx = np.argmax(j_scores)
optimal_threshold = thresholds[optimal_idx]
print(f"\nüéØ Optimal threshold: {optimal_threshold:.4f}")

# Recalculate predictions with optimal threshold
y_pred_optimal = (y_pred_prob >= optimal_threshold).astype(int)
cm_optimal = confusion_matrix(y_test, y_pred_optimal)
accuracy_optimal = accuracy_score(y_test, y_pred_optimal)

print("\nüìä Metrik dengan threshold optimal:")
print(f"Accuracy: {accuracy_optimal:.4f}")
print(f"Confusion Matrix:\n{cm_optimal}")
print(classification_report(y_test, y_pred_optimal))

# Business metrics
total_customers = len(y_test)
actual_churn = np.sum(y_test)
predicted_churn = np.sum(y_pred_class)
predicted_churn_optimal = np.sum(y_pred_optimal)

print("\nüíº Metrik Bisnis:")
print(f"Total pelanggan dalam test set: {total_customers}")
print(f"Pelanggan aktual yang churn: {actual_churn} ({actual_churn/total_customers:.2%})")
print(f"Pelanggan diprediksi churn (threshold=0.5): {predicted_churn} ({predicted_churn/total_customers:.2%})")
print(f"Pelanggan diprediksi churn (threshold optimal={optimal_threshold:.4f}): {predicted_churn_optimal} ({predicted_churn_optimal/total_customers:.2%})")

# Identify high-risk customers
hasil_df = pd.DataFrame({
    'id': idx_test,
    'actual_churn': y_test,
    'predicted_churn': y_pred_optimal.flatten(),
    'churn_probability': y_pred_prob.flatten().round(4)
})

# Sort by churn probability (highest first)
high_risk_customers = hasil_df.sort_values('churn_probability', ascending=False).head(10)
print("\n‚ö†Ô∏è 10 Pelanggan dengan risiko churn tertinggi:")
print(high_risk_customers)

# Feature importance analysis using a simple approach with permutation
print("\nüîç Feature Importance Analysis...")
feature_importances = []

# Calculate feature importance by permutation
# Save original features
baseline_score = accuracy_score(y_test, y_pred_class)
feature_names = list(X.columns) if isinstance(X, pd.DataFrame) else [f"feature_{i}" for i in range(X_scaled.shape[1])]

# Simple permutation importance implementation
for i in range(X_test.shape[1]):
    # Copy the test data
    X_test_permuted = X_test.copy()
    # Permute one feature
    X_test_permuted[:, i] = np.random.permutation(X_test_permuted[:, i])
    # Predict with permuted feature
    y_pred_permuted = (model.predict(X_test_permuted) > 0.5).astype(int)
    # Calculate new score
    permuted_score = accuracy_score(y_test, y_pred_permuted)
    # Feature importance is the decrease in performance
    importance = baseline_score - permuted_score
    feature_importances.append((feature_names[i], importance))

# Sort feature importance
feature_importances.sort(key=lambda x: x[1], reverse=True)

# Create DataFrame for visualization
fi_df = pd.DataFrame(feature_importances, columns=['feature', 'importance'])
print(fi_df.head(10))

# Visualize feature importance
plt.figure(figsize=(12, 8))
sns.barplot(x='importance', y='feature', data=fi_df.head(10))
plt.title('Top 10 Feature Importance')
plt.tight_layout()
plt.show()

# Save prediction results to CSV
hasil_df.to_csv("hasil_prediksi_churn_mlp_enhanced.csv", index=False)
print("\n‚úÖ Hasil prediksi telah disimpan ke 'hasil_prediksi_churn_mlp_enhanced.csv'")

# Save the model
model.save('telecom_churn_mlp_model_enhanced.h5')
print("‚úÖ Model telah disimpan ke 'telecom_churn_mlp_model_enhanced.h5'")

# Create a brief summary report
print("\nüìã Ringkasan Model Prediksi Churn (MLP Enhanced)")
print("=" * 50)
print(f"Jumlah total pelanggan dalam analisis: {len(df)}")
print(f"Jumlah total pelanggan churn: {df['churn'].sum()} ({df['churn'].mean():.2%})")
print(f"Akurasi model: {accuracy:.2%}")
print(f"AUC score: {roc_auc:.4f}")
print("\n‚úÖ Analisis Selesai!")
```

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from sklearn.utils.class_weight import compute_class_weight
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline

# Set random seeds untuk reproducibility
np.random.seed(42)

# Membaca dataset
print("üìä Memuat dataset...")
df = pd.read_csv('Dataset1.csv')

# Mengecek informasi dataset
print(f"\nüìã Informasi dataset: {df.shape[0]} baris √ó {df.shape[1]} kolom")
print("\nüîç Sampel data:")
print(df.head(5))

# Mengecek jumlah nilai NaN per kolom
print("\nüîé Jumlah NaN sebelum pengisian:")
print(df.isna().sum())

# Mengisi nilai NaN dengan rata-rata untuk kolom numerik
df.fillna(df.mean(numeric_only=True), inplace=True)

# Mengecek kembali jumlah NaN setelah pengisian
print("\n‚úÖ Jumlah NaN setelah pengisian:")
print(df.isna().sum().sum())

# Visualisasi distribusi target
plt.figure(figsize=(10, 6))
sns.countplot(x=df['churn'])
plt.title('Distribusi Churn', fontsize=16)
plt.xlabel('Churn Status', fontsize=14)
plt.ylabel('Jumlah Pelanggan', fontsize=14)
plt.tight_layout()
plt.show()

# Encode kolom target "churn" (Yes/No ‚Üí 1/0)
print("\nüîÑ Encoding target variable...")
le = LabelEncoder()
df['churn'] = le.fit_transform(df['churn'])
print(f"Label encoding: {le.classes_} ‚Üí {le.transform(le.classes_)}")

# Feature correlation analysis
print("\nüìà Analisis korelasi fitur dengan target...")
plt.figure(figsize=(12, 10))
correlation = df.corr()
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='coolwarm',
            linewidths=0.5, vmin=-1, vmax=1)
plt.title('Korelasi Antar Fitur', fontsize=16)
plt.tight_layout()
plt.show()

# Top features correlation with churn
churn_corr = correlation['churn'].drop('churn').sort_values(ascending=False)
print("\nTop korelasi dengan churn:")
print(churn_corr.head(5))

# Pisahkan fitur dan target
print("\nüîÑ Mempersiapkan data untuk pemodelan...")
X = df.drop("churn", axis=1)
y = df["churn"]

# One-hot encode kategori
categorical_cols = X.select_dtypes(include=['object']).columns
if len(categorical_cols) > 0:
    print(f"One-hot encoding kolom kategori: {list(categorical_cols)}")
    X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# Normalisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(f"Data dinormalisasi dengan StandardScaler. Shape: {X_scaled.shape}")

# Train-test split dengan mempertahankan index untuk tracking
X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
    X_scaled, y, df.index, test_size=0.3, random_state=42, stratify=y)

print(f"Train set: {X_train.shape[0]} sampel, Test set: {X_test.shape[0]} sampel")

# Class balancing
print("\n‚öñÔ∏è Menghitung class weights untuk mengatasi imbalance...")
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
class_weights_dict = {0: class_weights[0], 1: class_weights[1]}
print(f"Class weights: {class_weights_dict}")

# SVM Model dengan parameter yang dioptimalkan
print("\nüß† Membangun model SVM yang dioptimalkan...")
# Parameter yang dioptimalkan berdasarkan karakteristik dataset
model = SVC(
    kernel='rbf',          # Radial Basis Function kernel cocok untuk dataset ini
    C=10.0,                # Regularization parameter yang lebih tinggi untuk mengikuti pola data lebih dekat
    gamma=0.1,             # Gamma yang lebih besar memberikan decision boundary yang lebih fleksibel
    probability=True,      # Enable probability estimates
    class_weight=class_weights_dict,  # Apply class weights
    cache_size=1000,       # Menambah cache untuk performa lebih baik
    random_state=42        # For reproducibility
)

# Model training
print("\nüèãÔ∏è Melatih model SVM...")
model.fit(X_train, y_train)

print("‚úÖ Model SVM telah dilatih!")

# Evaluasi pada data test
print("\nüìä Evaluasi model pada test set...")
y_pred_class = model.predict(X_test)
y_pred_prob = model.predict_proba(X_test)[:, 1]  # Probabilitas kelas positif (1)

# Metrics calculation
accuracy = accuracy_score(y_test, y_pred_class)
cm = confusion_matrix(y_test, y_pred_class)

# Print detailed metrics
print("\nüìà Metrik Performa Model:")
print(f"Accuracy: {accuracy:.4f}")
print("\nüìä Confusion Matrix:")
print(cm)

print("\nüìù Classification Report:")
print(classification_report(y_test, y_pred_class))

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)
print(f"AUC: {roc_auc:.4f}")

# Enhanced visualization - Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Tidak Churn", "Churn"],
            yticklabels=["Tidak Churn", "Churn"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()

# Plot ROC Curve
plt.figure(figsize=(10, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()

# Calculate optimal threshold
j_scores = tpr - fpr
optimal_idx = np.argmax(j_scores)
optimal_threshold = thresholds[optimal_idx]
print(f"\nüéØ Optimal threshold: {optimal_threshold:.4f}")

# Recalculate predictions with optimal threshold
y_pred_optimal = (y_pred_prob >= optimal_threshold).astype(int)
cm_optimal = confusion_matrix(y_test, y_pred_optimal)
accuracy_optimal = accuracy_score(y_test, y_pred_optimal)

print("\nüìä Metrik dengan threshold optimal:")
print(f"Accuracy: {accuracy_optimal:.4f}")
print(f"Confusion Matrix:\n{cm_optimal}")
print(classification_report(y_test, y_pred_optimal))

# Business metrics
total_customers = len(y_test)
actual_churn = np.sum(y_test)
predicted_churn = np.sum(y_pred_class)
predicted_churn_optimal = np.sum(y_pred_optimal)

print("\nüíº Metrik Bisnis:")
print(f"Total pelanggan dalam test set: {total_customers}")
print(f"Pelanggan aktual yang churn: {actual_churn} ({actual_churn/total_customers:.2%})")
print(f"Pelanggan diprediksi churn (default): {predicted_churn} ({predicted_churn/total_customers:.2%})")
print(f"Pelanggan diprediksi churn (threshold optimal={optimal_threshold:.4f}): {predicted_churn_optimal} ({predicted_churn_optimal/total_customers:.2%})")

# Identify high-risk customers
hasil_df = pd.DataFrame({
    'id': idx_test,
    'actual_churn': y_test,
    'predicted_churn': y_pred_optimal,
    'churn_probability': np.round(y_pred_prob, 4)
})

# Sort by churn probability (highest first)
high_risk_customers = hasil_df.sort_values('churn_probability', ascending=False).head(10)
print("\n‚ö†Ô∏è 10 Pelanggan dengan risiko churn tertinggi:")
print(high_risk_customers)

# Save prediction results to CSV
hasil_df.to_csv("hasil_prediksi_churn_svm.csv", index=False)
print("\n‚úÖ Hasil prediksi telah disimpan ke 'hasil_prediksi_churn_svm.csv'")

# Save the model
import joblib
joblib.dump(model, 'telecom_churn_svm_model.pkl')
print("‚úÖ Model telah disimpan ke 'telecom_churn_svm_model.pkl'")

# Create a brief summary report
print("\nüìã Ringkasan Model Prediksi Churn")
print("=" * 50)
print(f"Jumlah total pelanggan dalam analisis: {len(df)}")
print(f"Jumlah total pelanggan churn: {df['churn'].sum()} ({df['churn'].mean():.2%})")
print(f"Akurasi model: {accuracy:.2%}")
print(f"AUC score: {roc_auc:.4f}")
print("\n‚úÖ Analisis Selesai!")
```
